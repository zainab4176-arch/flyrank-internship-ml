## 3. Quick look at the data (2-3 real numbers)

I loaded the starter dataset and examined a few basic statistics to understand whether this lane is worth exploring. The dataset contains **30,000 pages**, the **average search volume is approximately 158.88**, and **no pages have zero impressions** in this dataset. These observations show that the dataset is large enough and contains useful information to support a decision-support project for identifying pages that may benefit from content refresh.

```python
import os
import subprocess
import pandas as pd

# Clone the repository if it is not already present
if not os.path.exists("flyrank-ml-internship-starter"):
    subprocess.run([
        "git", "clone", "--depth", "1",
        "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
    ], check=True)

# Load the starter dataset
df = pd.read_csv(
    "flyrank-ml-internship-starter/data/raw/content_refresh_anonymized.csv"
)

# Calculate 3 real numbers
total_pages = len(df)
zero_impressions = (df["impressions_90d"] == 0).sum()
avg_search_volume = df["search_volume"].mean()

print(f"Total pages: {total_pages}")
print(f"Pages with zero impressions: {zero_impressions}")
print(f"Average search volume: {avg_search_volume:.2f}")
```
